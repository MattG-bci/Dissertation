```python
x = img_norm.copy() # creating copy to avoid modifying original images due to the pointer nature
uncalibrated_model = MPCATrainer(classifier='lr', n_features=200)
calibrated_model = MPCATrainer(classifier='lr', n_features=200)
calibrated_model.clf = CalibratedClassifierCV(calibrated_model.clf, cv=10, method='sigmoid')

calib_det_model = MPCATrainer(classifier='lr', n_features=200)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
X_val = X_train.copy()
y_val = y_train.copy()
```

```python
X_train.shape, X_val.shape, X_test.shape
```

```python
uncalibrated_model.fit(X_train, y_train)
calibrated_model.fit(X_train, y_train)
calib_det_model.fit(X_train, y_train)
```

```python
from scipy.optimize import minimize
from scipy.special import logit, expit

y_pred_binary = calib_det_model.predict_proba(X_val)
y_logits = logit(y_pred_binary)

def scale_parameters(x, *args):
    a, b = x
    y_logit_scaled = a*y_logits + b
    y_pred_inner = expit(y_logit_scaled)
    bce = sum([-(y_t * np.log(y_p) + (1 - y_t)*np.log(1 - y_p)) for y_t, y_p in zip(y_val, y_pred_inner[:, 1]) if not y_p==0])
    return bce

min_obj = minimize(scale_parameters, [1, 0], method='Nelder-Mead', options={'xatol': 1e-8, 'disp': True})
min_obj
```

```python
y_probs_calb = calib_det_model.predict_proba(X_test)
y_logits = logit(y_probs_calb)

y_test_pred_corr = expit(min_obj.x[0]*y_logits + min_obj.x[1])
```

```python
def selective_function(predictions, theta): #rejection function
    reject_idxs = []
    for index, prob in enumerate(predictions):
        if prob < theta:
            reject_idxs.append(index)
        
        else:
            continue
    
    return reject_idxs
```
